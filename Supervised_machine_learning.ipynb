```python
#Create dependent and independent variables
x = data[["Age (months)"]]#independent variable
y = data[["Child's weight (kilograms)"]]#dependent variable
```

```python
x
```

```python
y
```

```python
from sklearn  import linear_model
from sklearn.model_selection import train_test_split

model = linear_model.LinearRegression()

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

```

#Training the model


```python
model.fit(X_train,y_train)
```

#PREDICT the model

```python
y_pred = model.predict(X_test)
```

#Compare the actual vlaues with the predicted values

```python
plt.figure(figsize=(10,5))
plt.scatter(X_test,y_test,label = "Actual values", color = "red",marker=".")
plt.plot(X_test,y_pred,label = "Predicted values", color = "blue")
plt.legend()
plt.title("Actual vs Predicted values")
plt.xlabel("Age (months)")
plt.ylabel("Child's weight (kilograms)")
plt.show()
```
