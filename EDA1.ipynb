# Columns

####In Object-Oriented Programming, attributes and methods are key components of objects (like DataFrames in Pandas).

**Attributes:**

Represent the characteristics or properties of an object.
Store data associated with the object.
In Pandas, examples include df.shape (dimensions of the DataFrame), df.columns (column names), etc.

Methods:

Define the actions or behaviors that an object can perform.
Are functions associated with an object.
In Pandas, examples include df.head() (display first few rows), df.describe() (summary statistics), etc.
Key Difference: Attributes store data, while methods perform actions on the object.

For example:

df.shape is an attribute that stores the dimensions of the DataFrame.
df.sort_values() is a method that sorts the DataFrame based on specified columns.
Think of it like a car:

Attributes are like its color, number of seats, etc.
Methods are like starting the engine, accelerating, braking, etc.

In pandas, the pd.crosstab() function computes a cross-tabulation (also known as a contingency table) of two or more factors. It summarizes the relationship between these factors by showing the frequency distribution of each combination of factor levels.

```python
freq_table = pd.crosstab(df['Gender'], 'Bridget')
print(freq_table)
```

```python
freq_table = pd.crosstab(df['Education'],df['HHIncome'])
freq_table
```

```python
row_percentages = freq_table.apply(lambda row: row/row.sum(), axis=1)
print(row_percentages)
```

##Plotting pie chart of the education level
plt.pie(Education_counts, ...):** Creates a pie chart using the values from Education_counts.

**labels=Education_counts.index:** Sets the labels for each slice of the pie chart to the index (likely the education levels) of the Education_counts Series.

**autopct='%1.1f%%':** Formats the percentage labels on each slice to display one decimal place.

**startangle=90:** Rotates the pie chart so that the first slice starts at the 90-degree position (top).

```python
#Drawing a pie chart for education
Education_counts = df['Education'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(Education_counts, labels=Education_counts.index, autopct='%1.1f%%', startangle=40)
plt.title('Education Distribution')
plt.show()
```
